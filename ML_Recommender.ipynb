```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
movies_df.head()
```

```python
num_users = ratings_df['userId'].nunique()
num_rated_movies = ratings_df['movieId'].nunique()
total_ratings = len(ratings_df)
```

```python
rating_distribution = ratings_df['rating'].value_counts().sort_index()
```

```python
top_user = ratings_df['userId'].value_counts().idxmax()
top_user_ratings = ratings_df['userId'].value_counts().max()
```

```python
top_movie_id = ratings_df['movieId'].value_counts().idxmax()
top_movie_ratings = ratings_df['movieId'].value_counts().max()
```

```python
genre_counts = Counter(
    genre for genre_str in movies_df['genres'] for genre in genre_str.split('|')
)
genre_counts_df = pd.DataFrame(genre_counts.items(), columns=["Genre", "Count"])
genre_counts_df = genre_counts_df.sort_values(by="Count", ascending=False)
```

```python
sns.set(style="whitegrid", font_scale=1.1)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# a. Histogram Distribusi Rating
sns.histplot(ratings_df['rating'], bins=10, kde=False, ax=axs[0, 0], color="skyblue", edgecolor="black")
axs[0, 0].set_title("Distribusi Rating")
axs[0, 0].set_xlabel("Rating")
axs[0, 0].set_ylabel("Jumlah Rating")

# b. Bar Chart 10 Pengguna Paling Aktif
top_users = ratings_df['userId'].value_counts().head(10)
sns.barplot(x=top_users.index.astype(str), y=top_users.values, ax=axs[0, 1], palette="viridis")
axs[0, 1].set_title("10 Pengguna Paling Aktif")
axs[0, 1].set_xlabel("User ID")
axs[0, 1].set_ylabel("Jumlah Rating")

# c. Bar Chart 10 Film dengan Rating Terbanyak
top_movies = ratings_df['movieId'].value_counts().head(10)
# Gabungkan dengan judul film menggunakan movies_df
top_movie_titles = movies_df.set_index('movieId').loc[top_movies.index]['title']
sns.barplot(y=top_movie_titles, x=top_movies.values, ax=axs[1, 0], palette="magma")
axs[1, 0].set_title("10 Film dengan Rating Terbanyak")
axs[1, 0].set_xlabel("Jumlah Rating")
axs[1, 0].set_ylabel("Judul Film")

# d. Bar Chart Distribusi Genre (Top 10)
top_genres = genre_counts_df.head(10)
sns.barplot(y=top_genres["Genre"], x=top_genres["Count"], ax=axs[1, 1], palette="cubehelix")
axs[1, 1].set_title("Distribusi Genre Film (Top 10)")
axs[1, 1].set_xlabel("Jumlah Film")
axs[1, 1].set_ylabel("Genre")

plt.tight_layout()
plt.show()
```

```python
print("=== Ringkasan Analisis ===")
print(f"Jumlah user unik: {num_users}")
print(f"Jumlah film yang diberi rating: {num_rated_movies}")
print(f"Total rating: {total_ratings}")
print("\nDistribusi Rating:")
print(rating_distribution)
print(f"\nPengguna paling aktif: User {top_user} dengan {top_user_ratings} rating")
print(f"Film dengan rating terbanyak: '{top_movie_title}' (movieId: {top_movie_id}) dengan {top_movie_ratings} rating")
print("\nDistribusi Genre (Top 10):")
print(top_genres)
```

##STEAM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("steam-200k.csv", header=None, names=["user_id", "game", "behavior", "hours", "timestamp"])
df_play = df[df['behavior'] == 'play'].copy()
df_play['sim_timestamp'] = np.linspace(0, 1, len(df_play))
most_popular = df_play.groupby('game')['hours'].sum().sort_values(ascending=False).head(10)
```

```python
def linear_decay(t):
    return 1 - t

def exponential_decay(t):
    return np.exp(-5 * t)

def custom_decay(t):
    return 1 / (1 + 10 * t**2)
```

```python
df_play['linear_weight'] = linear_decay(df_play['sim_timestamp'])
df_play['exp_weight'] = exponential_decay(df_play['sim_timestamp'])
df_play['custom_weight'] = custom_decay(df_play['sim_timestamp'])
```

```python
def compute_trending(df, weight_col):
    df['weighted_hours'] = df['hours'] * df[weight_col]
    return df.groupby('game')['weighted_hours'].sum().sort_values(ascending=False).head(10)
```

```python
trending_linear = compute_trending(df_play.copy(), 'linear_weight')
trending_exp = compute_trending(df_play.copy(), 'exp_weight')
trending_custom = compute_trending(df_play.copy(), 'custom_weight')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 10))
axs = axs.flatten()

# Most Popular
axs[0].barh(most_popular[::-1].index, most_popular[::-1].values, color='skyblue')
axs[0].set_title('Most Popular Games (Total Hours Played)')

# Trending - Linear
axs[1].barh(trending_linear[::-1].index, trending_linear[::-1].values, color='lightgreen')
axs[1].set_title('Trending Games (Linear Decay)')

# Trending - Exponential
axs[2].barh(trending_exp[::-1].index, trending_exp[::-1].values, color='salmon')
axs[2].set_title('Trending Games (Exponential Decay)')

# Trending - Custom
axs[3].barh(trending_custom[::-1].index, trending_custom[::-1].values, color='orange')
axs[3].set_title('Trending Games (Custom Decay)')

plt.tight_layout()
plt.show()
```

##Books

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import hstack
```

```python
books = pd.read_csv("books.csv")
book_tags = pd.read_csv("book_tags.csv")
tags = pd.read_csv("tags.csv")
```

```python
book_tags_merged = pd.merge(book_tags, tags, on="tag_id", how="left")
tags_grouped = book_tags_merged.groupby('goodreads_book_id')['tag_name'].apply(lambda x: ' '.join(x.astype(str))).reset_index()
```

```python
books_metadata = pd.merge(books, tags_grouped, on='goodreads_book_id', how='left')
books_metadata['tag_name'] = books_metadata['tag_name'].fillna('')
```

```python
vectorizer_author = TfidfVectorizer(stop_words='english')
tfidf_matrix_author = vectorizer_author.fit_transform(books_metadata['authors'])

vectorizer_tags = TfidfVectorizer()
tfidf_matrix_tags = vectorizer_tags.fit_transform(books_metadata['tag_name'])
```

```python
tfidf_matrix_combined = hstack([
    tfidf_matrix_author * 0.4,
    tfidf_matrix_tags * 0.6
])
```

```python
def recommend_books_with_preferences(favorite_titles, method='combined', top_n=10):
    sim_scores = []
    for title in favorite_titles:
        idx = books_metadata[books_metadata['title'].str.lower() == title.lower()].index
        if len(idx) == 0:
            print(f" Buku dengan judul '{title}' tidak ditemukan.")
            continue
        idx = idx[0]

        if method == 'author':
            sim = cosine_similarity(tfidf_matrix_author[idx], tfidf_matrix_author).flatten()
        elif method == 'tags':
            sim = cosine_similarity(tfidf_matrix_tags[idx], tfidf_matrix_tags).flatten()
        elif method == 'combined':
            sim = cosine_similarity(tfidf_matrix_combined[idx], tfidf_matrix_combined).flatten()
        else:
            return " Metode tidak dikenal. Gunakan: 'author', 'tags', atau 'combined'."

        sim_scores.append(sim)

    if not sim_scores:
        return "Tidak ada buku yang ditemukan dalam preferensi."

    avg_sim_scores = np.mean(sim_scores, axis=0)
    recommended_indices = avg_sim_scores.argsort()[::-1]

    favorite_indices = [books_metadata[books_metadata['title'].str.lower() == title.lower()].index[0]
                        for title in favorite_titles if title.lower() in books_metadata['title'].str.lower().values]

    final_indices = [i for i in recommended_indices if i not in favorite_indices][:top_n]

    return books_metadata.iloc[final_indices][['title', 'authors']].reset_index(drop=True)
```

```python
favorite_books = ["Twilight", "The Hobbit", "The Catcher in the Rye"]
recommend_books_with_preferences(favorite_books, method='combined')
```

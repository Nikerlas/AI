```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
movies_df.head()
```

```python
num_users = ratings_df['userId'].nunique()
num_rated_movies = ratings_df['movieId'].nunique()
total_ratings = len(ratings_df)
```

```python
rating_distribution = ratings_df['rating'].value_counts().sort_index()
```

```python
top_user = ratings_df['userId'].value_counts().idxmax()
top_user_ratings = ratings_df['userId'].value_counts().max()
```

```python
top_movie_id = ratings_df['movieId'].value_counts().idxmax()
top_movie_ratings = ratings_df['movieId'].value_counts().max()
```

```python
genre_counts = Counter(
    genre for genre_str in movies_df['genres'] for genre in genre_str.split('|')
)
genre_counts_df = pd.DataFrame(genre_counts.items(), columns=["Genre", "Count"])
genre_counts_df = genre_counts_df.sort_values(by="Count", ascending=False)
```

```python
sns.set(style="whitegrid", font_scale=1.1)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# a. Histogram Distribusi Rating
sns.histplot(ratings_df['rating'], bins=10, kde=False, ax=axs[0, 0], color="skyblue", edgecolor="black")
axs[0, 0].set_title("Distribusi Rating")
axs[0, 0].set_xlabel("Rating")
axs[0, 0].set_ylabel("Jumlah Rating")

# b. Bar Chart 10 Pengguna Paling Aktif
top_users = ratings_df['userId'].value_counts().head(10)
sns.barplot(x=top_users.index.astype(str), y=top_users.values, ax=axs[0, 1], palette="viridis")
axs[0, 1].set_title("10 Pengguna Paling Aktif")
axs[0, 1].set_xlabel("User ID")
axs[0, 1].set_ylabel("Jumlah Rating")

# c. Bar Chart 10 Film dengan Rating Terbanyak
top_movies = ratings_df['movieId'].value_counts().head(10)
# Gabungkan dengan judul film menggunakan movies_df
top_movie_titles = movies_df.set_index('movieId').loc[top_movies.index]['title']
sns.barplot(y=top_movie_titles, x=top_movies.values, ax=axs[1, 0], palette="magma")
axs[1, 0].set_title("10 Film dengan Rating Terbanyak")
axs[1, 0].set_xlabel("Jumlah Rating")
axs[1, 0].set_ylabel("Judul Film")

# d. Bar Chart Distribusi Genre (Top 10)
top_genres = genre_counts_df.head(10)
sns.barplot(y=top_genres["Genre"], x=top_genres["Count"], ax=axs[1, 1], palette="cubehelix")
axs[1, 1].set_title("Distribusi Genre Film (Top 10)")
axs[1, 1].set_xlabel("Jumlah Film")
axs[1, 1].set_ylabel("Genre")

plt.tight_layout()
plt.show()
```

```python
print("=== Ringkasan Analisis ===")
print(f"Jumlah user unik: {num_users}")
print(f"Jumlah film yang diberi rating: {num_rated_movies}")
print(f"Total rating: {total_ratings}")
print("\nDistribusi Rating:")
print(rating_distribution)
print(f"\nPengguna paling aktif: User {top_user} dengan {top_user_ratings} rating")
print(f"Film dengan rating terbanyak: '{top_movie_title}' (movieId: {top_movie_id}) dengan {top_movie_ratings} rating")
print("\nDistribusi Genre (Top 10):")
print(top_genres)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
import pandas as pd

df = pd.read_csv("steam-200k.csv", header=None, names=["user_id", "game", "action", "value", "timestamp"])
df.head()
```

```python
df_playtime = df[df["action"] == "play"]
df_playtime["value"] = df_playtime["value"].astype(float)
df_playtime["timestamp"] = pd.to_datetime(df_playtime["timestamp"], unit="s")
```

```python
most_popular = df_playtime.groupby("game")["value"].sum().sort_values(ascending=False)
```

```python
from datetime import datetime

now = pd.Timestamp(datetime.utcnow())
df_playtime["days_ago"] = (now - df_playtime["timestamp"]).dt.days
```

```python
# Linear decay
def linear_decay(days_ago, max_days=365):
    return max(0, (max_days - days_ago) / max_days)

# Exponential decay
def exponential_decay(days_ago, alpha=0.05):
    return np.exp(-alpha * days_ago)

# Custom decay (misal: sigmoid-style decay)
def custom_decay(days_ago):
    return 1 / (1 + np.log1p(days_ago))
```

```python
import numpy as np

def trending_score(df, decay_func):
    df = df.copy()
    df["decay_weight"] = df["days_ago"].apply(decay_func)
    df["weighted_value"] = df["value"] * df["decay_weight"]
    return df.groupby("game")["weighted_value"].sum().sort_values(ascending=False)
```

```python
linear_scores = trending_score(df_playtime, linear_decay)
exp_scores = trending_score(df_playtime, exponential_decay)
custom_scores = trending_score(df_playtime, custom_decay)
```

```python
ranking_df = pd.DataFrame({
    "Most Popular": most_popular.rank(ascending=False),
    "Linear Decay": linear_scores.rank(ascending=False),
    "Exponential Decay": exp_scores.rank(ascending=False),
    "Custom Decay": custom_scores.rank(ascending=False)
}).dropna()
```

```python
import matplotlib.pyplot as plt

top_games = ranking_df.sort_values("Most Popular").head(10)

top_games[["Linear Decay", "Exponential Decay", "Custom Decay"]].plot(kind="bar", figsize=(12,6))
plt.title("Perbandingan Ranking Game Berdasarkan Pendekatan Decay")
plt.ylabel("Ranking")
plt.xlabel("Game")
plt.xticks(rotation=45, ha='right')
plt.gca().invert_yaxis()  # karena ranking lebih kecil = lebih tinggi
plt.tight_layout()
plt.show()
```

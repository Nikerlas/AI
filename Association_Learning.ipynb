```python
import pandas as pd
import numpy as np
import apyori as apriori
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
dataset.head()
```

```python
dataset.shape
```

```python
transaction=[]
for i in range(0,7501):
  transaction.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
transaction[:2]
```

```python
from apyori import apriori
rules=apriori(transaction,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

```python
results=list(rules)
results
```

```python
lhs,rhs,support,confidence,lift=[],[],[],[],[]
for result in results:
    lhs.append(tuple(result[2][0][0])[0])
    rhs.append(tuple(result[2][0][1])[0])
    support.append(result[1])
    confidence.append(result[2][0][2])
    lift.append(result[2][0][3])
```

```python
columns=list(zip(lhs,rhs,support,confidence,lift))
```

```python
result_df=pd.DataFrame(columns)
result_df.columns=['Item1','Item2','Support','Confidence','Lift']
result_df
```

```python
result_df=result_df.sort_values('Lift',ascending=False)
result_df
```

```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
dataset=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
dataset.head()
```

```python
dataset.shape
```

```python
transactions=[]
for i in range(0,7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
transactions[:2]
```

```python
from apyori import apriori
rules=apriori(transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

```python
results=list(rules)
results
```

```python
lhs,rhs,support=[],[],[]
for result in results:
    lhs.append(tuple(result[2][0][0])[0])
    rhs.append(tuple(result[2][0][1])[0])
    support.append(result[1])
```

```python
columns=list(zip(lhs,rhs,support))
```

```python
result_df=pd.DataFrame(columns)
result_df.columns=['Item1','Item2','Support']
result_df
```

```python
result_df=result_df.sort_values('Support',ascending=False)
result_df
```
